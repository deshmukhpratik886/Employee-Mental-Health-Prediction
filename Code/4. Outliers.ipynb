### 4.3 Checking the Outliers, MisMatch Values and Make it proper from Gender column

```python
# pick the gender col

# Gender column consist of lots of inconsistent values 

# printing unique values for gender col

print(df_final_merged_data['Gender'].unique())
```

```python
# Changing all values to lowercase

df_final_merged_data["Gender"] = df_final_merged_data["Gender"].str.lower()
```

Show the unique and mismatch values

```python
print(df_final_merged_data['Gender'].unique())
```

**Group the all mismatch values**

```python
# Making gender-groups to classify all values


male_str = ["male", "m", "male-ish", "maile", "mal", "male (cis)", "make", "male ",
            "man","msle", "mail", "malr","cis man", "Cis Male", "cis male",'male.',
            'nb masculine','sex is male','mtf',"i'm a man why didn't you make this a drop down question. you should of asked sex? and i would of answered yes please. seriously how much text can this take? ",
            'm|','cis-male','cis hetero male',"male (hey this is the tech industry you're talking about)",
            'male, cis','ostensibly male','male, born with xy chromosoms',
            'malel','cisgender male','let\'s keep it simple and say "male"',
            'identify as male','masculine','cishet male','i have a penis',
            'masculino','male/he/him','cis-het male','cis male ' ]



trans_str = ["trans-female", "something kinda male?", "queer/she/they",
             "non-binary","nah", "all", "enby", "fluid", "genderqueer", "androgyne", 
             "agender", "male leaning androgynous", "guy (-ish) ^_^", "trans woman", "neuter", 
             "female (trans)", "queer", "ostensibly male, unsure what that really means",
             'bigender','other/transfeminine','androgynous','genderfluid','nonbinary','female-bodied; no feelings about gender',
             'transgender woman','uhhhhhhhhh fem genderqueer?','agender/genderfluid','non binary',
             'genderqueer demigirl','genderqueer/non-binary','transfeminine',
             'male (or female, or both)','trans man','transgender','female/gender non-binary.',
             'demiguy','trans female','she/her/they/them','other',
             'nonbinary/femme','gender non-conforming woman','agender trans woman',
             'trans non-binary/genderfluid','non-binary and gender fluid','mostly male',
             'afab non-binary','non-binary/agender','male/genderqueer',
             'male (trans, ftm)','male/androgynous ' ]           




female_str = ["cis female", "f", "female", "woman",  "femake", "female ","cis-female/femme", 
              "female (cis)", "femail",'i identify as female.','female assigned at birth ',
              'fm','cis female ','genderfluid (born female)','female/woman','female or multi-gender femme',
              'cisgender female','genderqueer woman','fem','female (props for making this a freeform field, though)',
              ' female','cis-woman','genderflux demi-girl','femalw','my sex is female.','female (cisgender)','female (cis) ',
              'woman-identified',
              'cis-female','female-ish','i identify as female','*shrug emoji* (f)','cis woman',
                 "cisgendered woman",'female-identified','femmina','femile','female, she/her' ]
```

**Replace the mismatch value in meaningfull data**

```python
# Changing groups into three categories

for (row, col) in df_final_merged_data.iterrows():

    if (col.Gender) in male_str:
        df_final_merged_data['Gender'].replace(to_replace=col.Gender, value='male', inplace=True)

    if (col.Gender) in female_str:
        df_final_merged_data['Gender'].replace(to_replace=col.Gender, value='female', inplace=True)

    if (col.Gender) in trans_str:
        df_final_merged_data['Gender'].replace(to_replace=col.Gender, value='trans', inplace=True)

        
# Getting rid of unnnecessary values

stk_list = ['a little about you', 'p','none of your business','human','dude',
            'unicorn','cisdude','afab','god king of the valajar',
            'sometimes','contextual','\\-','none','swm','nb','questioning','rr',
            '43','i am a wookie','b','homem cis','transitioned, m2f',
            'male 9:1 female, roughly', 'f, cisgender']

df_final_merged_data = df_final_merged_data[~df_final_merged_data['Gender'].isin(stk_list)]
```

### 4.4 Handling Mismatch data from No_of_Employees column

```python
# now pick No_of_Employees col and remove or replace outliers

# first check the unique values in this col

df_final_merged_data['No_of_Employees'].unique()
```

**Grouping the mismatch data**

```python
# Making groups to classify all values

jun_str = ["Jun-25"]
may_str = ["01-May"]
```

**Change in Meaningfull data**

```python
# Changing groups into two categories

for (row, col) in df_final_merged_data.iterrows():

    if (col.No_of_Employees) in jun_str:
        df_final_merged_data['No_of_Employees'].replace(to_replace=col.No_of_Employees, value='6-25', inplace=True)

    if (col.No_of_Employees) in may_str:
        df_final_merged_data['No_of_Employees'].replace(to_replace=col.No_of_Employees, value='1-5', inplace=True)
```

### 4.5 Make data in proper and Meaningfull 

```python
# now pick Self_Employed col and remove or replace outliers

# first check the unique values in this col


df_final_merged_data['Self_Employed'].unique()
```

**Grouping the data**

```python
# Making groups to classify all values

zero_str = ["0"]
one_str = ["1"]
```

**Change in Proper and Meaningfull data**

```python
# Changing groups into two categories

for (row, col) in df_final_merged_data.iterrows():

    if (col.Self_Employed) in zero_str:
        df_final_merged_data['Self_Employed'].replace(to_replace=col.Self_Employed, value='False', inplace=True)

    if (col.Self_Employed) in one_str:
        df_final_merged_data['Self_Employed'].replace(to_replace=col.Self_Employed, value='True', inplace=True)
```

```python
# Values after cleaning

df_final_merged_data['Self_Employed'].value_counts()
```

```python
# now pick Self_Employed col and remove or replace outliers

# first check the unique values in this col


df_final_merged_data['Self_Employed'].unique()


# Making groups to classify all values

no_str = ["No"]
yes_str = ["Yes"]
```

```python
# Changing groups into two categories

for (row, col) in df_final_merged_data.iterrows():

    if (col.Self_Employed) in no_str:
        df_final_merged_data['Self_Employed'].replace(to_replace=col.Self_Employed, value='False', inplace=True)

    if (col.Self_Employed) in yes_str:
        df_final_merged_data['Self_Employed'].replace(to_replace=col.Self_Employed, value='True', inplace=True)
```
